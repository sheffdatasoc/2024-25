## Setup
### Load Iris and add Average Sepal Length

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')
avg_sepal_length = iris.groupby('species')['sepal_length'].mean().reset_index()
```

## Challenge 1: Basic Plots
### Objectives: 
-  Create a basic Line Plot
- Create a basic Bar Chart
- Add Labels, Title and Grid


```python
## Write your answers here
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x='species', y='petal_length', data=iris, marker="o", ci=None)
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='petal_length', data=iris)
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```

## Challenge 2

### Objective: 
- Create a Histogram
- Is it skewed? 
- What Statistics can you draw from it?


```python
## Write your answers here
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(iris['sepal_length'], bins=10, kde=True)
plt.title('Distribution of Sepal Lengths')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

# Challenge 3

```python
## Write your answers here

c=iris['species'].astype('category').cat.codes #How to encode for colour
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=iris['species'].astype('category').cat.codes, cmap='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.colorbar(ticks=[0, 1, 2], label='Species (0: Setosa, 1: Versicolor, 2: Virginica)')
plt.show()
```

## Challenge 4: 
### Ojectives:
-  Create a Box Plot
- Add a palette, a grid, detail
- Can you tell what the outliers are from this?


```python
## Write your answers here:
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set2')
plt.title('Petal Length Distribution by Species', fontsize=16)
plt.xlabel('Species', fontsize=14)
plt.ylabel('Petal Length', fontsize=14)
plt.grid(True)
plt.show()
```

## Challenge 5:

```python
plt.figure(figsize=(8, 5))
# Calculate the correlation matrix for only numeric features
correlation = iris.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

## Challenge 6:

### Objectives:
- Use subplots
- Show both the scatter plot and boxplot side by side
- Use a tight layout


```python
## Write your answer here
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot
sns.scatterplot(ax=axs[0], x='sepal_length', y='sepal_width', hue='species', data=iris)
axs[0].set_title('Sepal Length vs Sepal Width')

# Box plot
sns.boxplot(ax=axs[1], x='species', y='petal_length', data=iris)
axs[1].set_title('Petal Length by Species')

plt.tight_layout()
plt.show()
```
